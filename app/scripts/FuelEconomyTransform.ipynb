```python
# Imports & Pre-cleaning

import pandas as pd
import json

FuelData = pd.read_csv("../src/data/Misc/vehicles.csv")

#Subset to the useful columns
FuelData = FuelData[['make', 'model', 'year', "comb08", "fuelType", "fuelType1", "fuelType2", "cylinders", "trany"]]
```

```python
subsetFuelData = FuelData[['make', 'model','year', 'cylinders', 'trany', 'comb08']]
subsetFuelData = subsetFuelData[subsetFuelData.duplicated(keep=False)]
```

```python
# Extract unique Makes for React Selector
Makes = [{'value': m, 'label': m} for m in FuelData['make'].unique()]

Makes = pd.DataFrame(Makes)
```

```python
# Extract unique Models for React Selector conditioned on Make
Models = FuelData.groupby('make')['model'].unique().apply(
    lambda x: [{'value': model, 'label': model} for model in x]
).to_dict()
```

```python
# Export lists for react selectors
Makes.to_json('../src/data/Misc/MakeOptions.json', orient='records')

with open('../src/data/Misc/ModelOptions.json', 'w') as json_file:
    json.dump(Models, json_file)
```

```python
# Export dataframe for searching
FuelData.to_json('../src/data/Misc/FuelData.json', orient='records')
```
