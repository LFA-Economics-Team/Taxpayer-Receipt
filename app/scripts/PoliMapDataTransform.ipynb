```python
### Imports

import pandas as pd
import geopandas as gpd
import hvplot.pandas as hv
import geoviews as gv
import json

HouseShape = gpd.read_file("../src/data/Geospacial/UtahHouseDistricts2022to2032.shp")
SenateShape = gpd.read_file("../src/data/Geospacial/UtahSenateDistricts2022to2032.shp")

HouseShape = HouseShape.set_crs(epsg=3857)
SenateShape = SenateShape.set_crs(epsg=3857)
```

```python
### Output to usable JSON file

HouseShape.to_crs(epsg=4326).to_file('../src/data/Geospacial/HouseDistricts.json', driver='GeoJSON')
SenateShape.to_crs(epsg=4326).to_file('../src/data/Geospacial/SenateDistricts.json', driver='GeoJSON')
```
